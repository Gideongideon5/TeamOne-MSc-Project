```python
import time
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from datetime import datetime, timedelta, date
# Loading datasets
data = pd.read_csv("./chicago-traffic-tracker-historical-congestion-estimates-by-segment-2018-current-1.csv", na_values = "NaN")
```

```python
# Replacing -1 values (which is missing values) to NaN
data.replace(-1, 'NaN', inplace=True)
data = data[(data != "NaN")]
```

```python
#Deleting Unuseful Columns
del data['COMMENTS']
del data['Community Areas']
del data['Wards']
```

```python
# Before removing NaN Values the Dateset shape
data.shape
```

```python
#Describing Percentage for each column Total Missing Values in the Datasets
data.isnull().sum() * 100 / len(data)
```

```python
# NaN Values Drop and Clean Dataset 
data = data.dropna()
```

```python
# After removing NaN Values the Dateset shape
data.shape
```

```python
#Describing Percentage for each column Total Missing Values in the Datasets
data.isnull().sum() * 100 / len(data)
```

```python
#Extraction in the new column Year, Month, Day, Time from Date
data['TIME'] = pd.to_datetime(data['TIME'])
data['Date'] = data['TIME'].dt.date
data['Year'] = data.TIME.dt.year.astype('int64')
data['Month'] = data.TIME.dt.month.astype('int64')
data['Week'] = data.TIME.dt.week.astype('int64')
data['Day'] = data.TIME.dt.day.astype('int64')
```
